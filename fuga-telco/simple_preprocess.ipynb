```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
print("shape:", df.shape)
# replace space by np.nan
df.loc[df['PagoTotal'] == " ", 'PagoTotal'] = np.nan
# convert to float
df['PagoTotal'] = df['PagoTotal'].astype(float)
df.head()
```

```python
df.describe().round(2)
```

```python
# eliminamos valores outliers
df.loc[df.Antiguedad < 0, 'Antiguedad'] = np.nan
df.loc[df.PagoTotal > 8000, 'PagoTotal'] = np.nan
```

```python
print(df.isna().sum().sum()) # solo 122 NaN, omitiremos los valores perdidos
df = df.dropna()
```

```python
# reemplazamos los valores mas comunes
df = df.replace(to_replace='Si', value=1)
df = df.replace(to_replace='No', value=0)
df = df.replace(to_replace='No internet service', value=-1)
# dict TipoContrato
df = df.replace({'TipoContrato':{'2ano':2,'1ano':1,'Mensual':0}})
# dict Sexo
df = df.replace({'Sexo':{'F':0,'M':1}})
# dict Internet
df = df.replace({'Internet':{'Fibra Optica':1,'DSL':2}})
# dict MetodoPago
df = df.replace({'MetodoPago':{'pago online':0,'boleta':1,'CC':2,'TC':3}})
df.head()
```

```python
df.describe().round(1)
```

```python
df.to_csv("Telco-Customer-Churn-SP.csv", index=False)
```
